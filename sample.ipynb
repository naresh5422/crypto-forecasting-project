```python
import os
import shutil
import kagglehub
```

```python
# Download latest version
path = kagglehub.dataset_download("sudalairajkumar/cryptocurrencypricehistory")
print("Path to dataset files:", path)
```

```python
# Current working directory
cwd = os.getcwd()
# Target folder in current directory
target_path = os.path.join(cwd, "data")
# Copy dataset
shutil.copytree(path, target_path, dirs_exist_ok=True)
print("Dataset copied to:", target_path)
```
